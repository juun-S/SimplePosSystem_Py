```python
#POSsystem.py
menu = {"짜장면":5000,"짬뽕":6000,"볶음밥":6000,"탕수육":10000,} #메뉴판의 딕셔너리로 작성
table = {} 
table_order = {}
to =open("today_order.txt",'w') #결제내역을 저쟝할 txt파일 "w"로 생성
to.write("오늘의 결제내역입니다.") 
to.close() 

def main(): #포스프로그램의 메인함수정의     
    while True:
        main_mode_choose=input("""환영합니다.무인pos시스템입니다.
        메인모드를 선택해주세요.
        ==1.주문 및 결제모드==
        ==2.관리자모드==
        ==3.결제내역==
        ==종료==
        """)
        while True:
            while main_mode_choose=="1": 
                Order_mode() #주문결제모드가 끝나면 메인모드를 다시고르기위해 break시킨다
                break
                

            while main_mode_choose=="2":
                manager_mode() #2번선택시 관리자모드 실행
                break
            if main_mode_choose=="3":
                    orderlist() #3번선택시 오늘결제내역 txt로출력
                    break
            else:
                break
        if main_mode_choose=="종료":
            print("종료합니다.")
            break
        
                
#main()을 실행하면 환영한다는 말과함께 1번모드와 2번모드를 정의하고 
#다른 어떤 것을 입력했을 때 다시 input으로 모드를 입력하도록 whlie True를 사용
            
            
        
            


            

def Order_mode(): #main()모드에서 실행 할 수 있도록 주문모드Order_mode정의 
    table = {} 
    table_order = {}
    menu_sum = 0
    other_order_choose="1"
    if other_order_choose=="1": #주문을 다하고 난 뒤 1번을 선택하면 다시 주문모드를 실행하고
        pass
    else: #주문을 다하고 난 뒤 1번을 선택하지않으면 포스메인모드로 돌아갑니다.
        main()
    
    answer=input("환영합니다.주문하시겠습니까?예or아니요.:") #메뉴를 보여주고 주문의사여부를 입력합니다.
    if answer=="아니요":
        Order_mode() 
    while True: #while True 문으로 메뉴추가를 원할때 계속 실행되게합니다.
        while answer=="예":
            print(menu)
            order_menu = input("메뉴를 선택하세요.")
            order_quantity = int(input("주문하실 수량을 입력하세요.")) 
            menu_error= menu.get(order_menu,'메뉴가 존재하지 않습니다')
            while menu_error=='메뉴가 존재하지 않습니다':
                order_menu=input("메뉴를 다시입력하세요.")
                menu_error= menu.get(order_menu,'메뉴가 존재하지 않습니다') 
                continue #메뉴가 존재하지않으면 .get(,)문으로 기본값을 보여주고 존재하는 메뉴를 입력할 때까지 입력시킨다.
            while menu_error!='메뉴가 존재하지 않습니다':
                menu_sum = menu_sum + menu.get(order_menu) * order_quantity
                table_order[order_menu]=order_quantity
                answer = input("메뉴를 추가 하시겠습니까? 예 or 아니요 : ")
                break #menu안의 메뉴를 잘 입력하면 키값을 연산하고 메뉴추가주문의사를 input시킨다.
        
            
        
        
        if answer=="아니요":
            print("결제 금액은", menu_sum, "원 입니다.")
            break
             #추가주문이 없을 경우 주문가격총합을 print하고 결제로 넘어간다.
        if answer=="X":#추가주문의사가 없고 주문의사도 없을 시 메인으로 가기위한 키
            main()
            break 
        else:#예, 아니요가 아니라 나갈의사를 입력할 수 있는 기능 구현
            answer=input("""'예or아니요'로만 대답해주세요.주문하시겠습니까?
            (메인시스템으로 나가고싶다면"X"를 입력하세요.)
            """) 
            continue
                      

    
      
            
    if menu_sum>0:  #주문금액이 0보다 많아야 주문한 것이므로 >0으로 실행제약, 결제는 한번만실행하므로if문
        today_order=[table, menu_sum] #결제내역에 테이블주문내역과 총액 리스트로 저장
        table[input('테이블1,2,3,4중 하나를 입력해주세요.')]= table_order #테이블선택입력 딕셔너리화
        print(table) #주문확인
        order_answer=input("결제 하시겠습니까? 예 or 아니요.") #결제의사input
        while order_answer=="예": #예 일때만 실행 
            client_give = input('카드?현금?: ')
            if client_give =="카드": #카드로지불하면 바로 결제완료 
                print("계산 되었습니다.")
                to=open("today_order.txt",'a') #결제완료 되었으니 today_order.txt에 추가저장
                to.write(str(today_order))
                to.close()
                today_order=[table, menu_sum]
                break
            elif client_give=="현금":
                client_give_money = int(input("지불한 금액을 입력하세요."))
                order_change = client_give_money - menu_sum
                if order_change>=0: #거스름돈이 0이상이어야 거슬러주는 실행제약 
                    print('거스름돈은', order_change ,"입니다.")
                    to=open("today_order.txt",'a') #거스름돈이 0이상이어야결제된것이므로 if문안에서 저장실행
                    to.write(str(today_order))
                    to.close()
                    today_order=[table, menu_sum]
                    break
                else: 
                    print('지불하신금액이', order_change, '부족합니다')
            else:#결제의사가 없었거나 다른말을 입력시 다시 결제의사를 묻고 "예"이면 결제실행 
                print("다시 입력하세요.")
                order_answer= input("결제하시겠습니까? 예 or 아니요:")

    print("이용해주셔서 감사합니다.")
    other_order_choose=input("""다른주문을 원하시면 1 을
    종료하고싶으면 1을 제외한 다른 아무키나누르세요.""")
    if other_order_choose=="1":  #재주문의사에의한 모드실행을 위한제어
        Order_mode()  #재주문의사가 있으면 다시 주문모드실행
    else: #아니면 pass로 주문결제모드종료
        pass    
    

def manager_mode(): #메뉴판을 관리하기위해 main()에서 관리자모드를 들어오기위해 함수를 정의 
        print("""====관리자모드에 오신 것을 환영합니다====
        =========보안해제가 필요합니다.=========
        """)

        PASSWORD_input=input("암호를 입력해주세요.:")

        while PASSWORD_input!='password': #패스워드는 password로 고정,비번수정기능없음
            print("잘못된 암호입니다.")
            PASSWORD_input=input("다시 암호를 입력해주세요.:") #암호가 같지않으면 while문으로 루프안에 가둠

        if PASSWORD_input=='password':
            print("""===관리자님 반갑습니다.보안이 해제되었습니다.===\n
            ====모드를 실행하시겠습니까? 예 or아니요.====\n
            """)
            manager_answer=input("")
            while True: #3을 넣어서 모드를 종료시키기전에 종료되지 않기 위해 while True사용 
                if manager_answer=="예": #if문으로 관리자모드실행여부를 입력값에 따른 실행
                    print("""모드1: 메뉴와 가격추가모드
                    모드2: 메뉴수정 및 삭제모드
                    모드3: 나가기
                    """)
                elif manager_answer=="아니요": 
                        break
                else: #예아니요가 아닌 입력값이 들어올시 반복실행하기위한 제어
                    manager_answer=input("모드를 실행하시겠습니까?예or아니요")
                    continue
                choose_mode=input("모드번호를 입력하세요.:")
                while True: #while True로 원할 때 종료가능
                    while choose_mode=="1":
                        print("'.메뉴와 가격 추가모드'가 실행되었습니다.")
                        print("현재 활성된 메뉴:", menu)
                        print("추가를 원하시는 메뉴와 가격을 입력하세요.")
                        new_menu=input("메뉴:")
                        new_menu_price=int(input("가격:")) #모든 돈과 관련된 입력값은 연산을위해 int
                        menu[new_menu]=new_menu_price
                        print("새로운 메뉴와 가격은", new_menu, new_menu_price, "입니다.")
                        print("현재 활성화된 메뉴는", menu)
                        choose_mode=input("""원하는 모드를 선택하세요.
                        1.메뉴와 가격 추가모드
                        2.메뉴 수정 및 삭제모드
                        3.그만두기
                        """)
                    while choose_mode=="2":
                        print("'메뉴수정 및 삭제모드'가 실행되었습니다.")
                        print("현재 활성화된 메뉴:", menu) 
                        print("수정하거나 삭제할 메뉴와 가격을  입력하세요.")
                        edit_menu=input("메뉴:")
                        while str(edit_menu in menu) == "False":  
                            edit_menu=input("메뉴판 안의 메뉴만 입력하세요.")
                            continue #딕셔너리안의 key를 입력했는지 확인위해 in으로 확인 후 str로 바꿔제어
                            #그후 다시입력유도, key가 맞지않으면 continue로 반복

                        choose_edit_mode=input("수정하시려면 1을 입력하시고, 삭제하시려면 2를 입력하세요.")
                        while choose_edit_mode=="1":
                            del menu[edit_menu] #del로 수정할 메뉴{키:벨류}삭제 
                            edit_menu=input("*메뉴이름을 입력하세요:") 
                            edit_menu_price=int(input("*메뉴가격을 입력하세요.:"))
                            menu[edit_menu]=edit_menu_price #딕셔너리[키]=벨류 로 새로운 메뉴와가격입력 
                            print("메뉴를 수정하였습니다. 수정된 메뉴판은", menu, "입니다.")
                            choose_mode=input("""원하는 모드를 선택하세요.
                        1.메뉴와 가격 추가모드
                        2.메뉴 수정 및 삭제모드
                        3.그만두기
                        """)
                            break
                            if choose_mode=="3": 
                                menu[edit_menu]=edit_menu_price
                                break
                        while choose_edit_mode=="2":
                            print("삭제할 메뉴를 입력하세요.")
                            del menu[edit_menu]
                            print("메뉴를 삭제하였습니다.")
                            print("현재 활성화된 메뉴판은", menu,"입니다.")
                            choose_mode=input("""원하는 모드를 선택하세요.
                        1.메뉴와 가격 추가모드
                        2.메뉴 수정 및 삭제모드
                        3.그만두기
                        """)
                            return menu
                            if choose_mode=="3":
                                menu[edit_menu]=edit_menu_price
                                break
                    if choose_mode=="3":
                        print("그만둡니다.")
                        menu[edit_menu]=edit_menu_price
                        break
                    else:
                        choose_mode=input("""원하는 모드를 선택하세요.
                        1.메뉴와 가격 추가모드
                        2.메뉴 수정 및 삭제모드
                        3.그만두기
                        """)
                        continue
                        
                        
                if manager_answer=="아니요":
                        menu[new_menu]=new_menu_price #수정이끝나고 변경된것들을 전체적으로 menu에 입력
                        menu[edit_menu]=edit_menu_price
                        break
                if choose_mode=="3":
                    break
        return menu #관리된 메뉴판을 다른모드에서도 사용가능하게 return
    #당일 주문 내역 만들기
def orderlist():  
      with open("today_order.txt",'r')as to: #결제내역 인출을 위해 r로 읽기모드로 실행
            order_list=to.readlines() #모든 결제내역을 읽어온다
            for line in order_list: #라인한줄한줄 읽고
                print(line)#프린트
            
        
    
     
    
    
```
